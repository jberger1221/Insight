```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data[data.]
```

```python
data = pd.read_csv('../util/reviews_token.csv')
```

```python
data = pd.read_csv('../util/clean_reviews.csv')
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data.dropna(inplace=True)
```

```python
chart = sns.countplot(data.Question)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
```

# Cleaning pipeline

```python
pipeline = Pipeline([
    ('bow', CountVectorizer()),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
msg_train, msg_test, label_train, label_test = train_test_split(data.clean_answers, data.Question, test_size=0.4)
```

```python
pipeline.fit(msg_train,label_train)
```

```python
predictions = pipeline.predict(msg_test)
```

```python
print(confusion_matrix(label_test,predictions))
print('\n')
print(classification_report(label_test,predictions))
```
