```python
#Function to remove 
def convert_to_percent(val):
    if val == 0:
        return 0
    elif '/' in val:
        t = val.split('/')
        t = [int(i) for i in t]
        if t[0] == 0:
            return 0
        else:
            return 100*t[0]/t[1]
    else:
        return float(val)
```

```python
# Function to convert string to float
def convert_to_num (val):
    if ',' in val:
        val= ''.join(char for char in val if char != ',')
        return float(val)
    else:
        return float(val)
```
