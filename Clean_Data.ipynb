```python
#Function to remove 
def convert_to_num(val):
    if type(val) == int or type(val)==float:
        return val
    
    elif '/' in val:
        t = val.split('/')
        t = [int(i) for i in t]
        if t[0] == 0:
            return 0
        else:
            return 100*t[0]/t[1]
        
    elif '<' in val:
        val= ''.join(char for char in val if char != '<')
        val= ''.join(char for char in val if char != '%')
        return float(val)   
    
    elif '%' in val:
        val= ''.join(char for char in val if char != '%')
        return float(val)   
    
    elif ',' in val:
        val= ''.join(char for char in val if char != ',')
        return float(val)
    
    else:
        return float(val)
```

```python
def convert_col_to_num (dataf):
    for i in range (2,(len(dataf.columns))):
        dataf[dataf.columns[i]] = dataf[dataf.columns[i]].apply(convert_to_num)
    return dataf
```
