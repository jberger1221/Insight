```python
def extract_reviews(url,Nn,M,df_patients):
    j=0
    reviews_list = ""
    textr="" 
    print(url)
    page = requests.get(url)
    html=page.text   
    
    soup = BeautifulSoup(html,'html.parser')
    reviews=soup.find_all(class_='review-snippet')
    
    if (reviews):
           score=soup.find(class_="nps-badge nps-badge--clinic").get_text()
    num=0
    
    for r in reviews:
            age=r.find(class_='reviewer-detail reviewer-detail--age').get_text()[3:]
            treatment=r.find(class_='reviewer-detail reviewer-detail--treatment').get_text()[9:]
            success=r.find(class_="reviewer-detail reviewer-detail--success-with-this-doc").get_text()[21:]
            procedure=r.find(class_='reviewer-detail reviewer-detail--treatment').get_text()[9:]
            diagnosis=r.find(class_="reviewer-detail reviewer-detail--diagnosis").get_text()[9:]
            df_patients.loc[M]=[Nn]+ [url]+[score] + [age]+  [treatment]+[success]+ [diagnosis]+[procedure]
            M=M+1
                  
    return df_patients, M

df_patients = pd.DataFrame(columns = ['Nn','url','score','age','treatment','success','diagnosis','procedure'])
i=1
                  

xls = pd.ExcelFile('C:/Insight/my_project/fertilityIQ.xls')
allclinics= pd.read_excel(xls)

Nn=0
M=0
    
for i in range(0,450):
    Nn=Nn+1
    url=allclinics.iloc[i,3]
    if (url!=0):
       df_patients, M = extract_reviews(url,Nn,M,df_patients)
       

df_patients.to_csv('reviews_score_fertility.csv') 
```
