```python
def concat_reviews(df):
    df_final = pd.DataFrame()
    
    list_of_topics = ['Doctor','Advice','Nurse','Clinic','Protocol','Cost','Problems']
    topic_list =[]
    for t in list_of_topics:
        for i in range(3):
            topic_list.append(t)
            
    for clinic in df.clinic_name.unique():
        #print(clinic)
        
        clinic_name = []
        for i in range(21):
            clinic_name.append(clinic)
        

        
        df_temp = df[df.clinic_name == clinic]
        
        #doctor=[]
        #advice=[]
        #nurse = []
        #clinic = []
        #protocol = []
        #cost = []
        #problem = []
        
        doctor_pos=[]
        doctor_neg=[]
        doctor_neu=[]
        
        advice_pos=[]
        advice_neg=[]
        advice_neu=[]
        
        nurse_pos=[]
        nurse_neg=[]
        nurse_neu=[]
        
        clinic_pos=[]
        clinic_neg=[]
        clinic_neu=[]
        
        protocol_pos=[]
        protocol_neg=[]
        protocol_neu=[]
        
        cost_pos=[]
        cost_neg=[]
        cost_neu=[]
        
        problem_pos=[]
        problem_neg=[]
        problem_neu=[]
        
        
        for q in df_temp.Question.unique():
            
            df_question = df_temp[df_temp.Question == q]
            
            
            if q == "Doctor":
                #doctor = df_question.Answer.sum()
                
                for sen in df_question.Sentiment.unique():
                    
                    df_sen = df_question[df_question.Sentiment == sen]
                    
                    if sen == 'POSITIVE':
                        doctor_pos = df_sen.Answer.sum()
                    elif sen == 'NEGATIVE':
                        doctor_neg = df_sen.Answer.sum()
                    else:
                        doctor_neu = df_sen.Answer.sum() 
                    
                
                
            elif q == "Advice":
                #advice = df_question.Answer.sum()
                
                for sen in df_question.Sentiment.unique():
                    
                    df_sen = df_question[df_question.Sentiment == sen]
                    
                    if sen == 'POSITIVE':
                        advice_pos = df_sen.Answer.sum()
                    elif sen == 'NEGATIVE':
                        advice_neg = df_sen.Answer.sum()
                    else:
                        advice_neu = df_sen.Answer.sum() 
                
            elif q == "Nurse":
                #nurse = df_question.Answer.sum()
                
                for sen in df_question.Sentiment.unique():
                    
                    df_sen = df_question[df_question.Sentiment == sen]
                    
                    if sen == 'POSITIVE':
                        nurse_pos = df_sen.Answer.sum()
                    elif sen == 'NEGATIVE':
                        nurse_neg = df_sen.Answer.sum()
                    else:
                        nurse_neu = df_sen.Answer.sum() 
                
            elif q == "Clinic":
                #clinic = df_question.Answer.sum()
                
                for sen in df_question.Sentiment.unique():
                    
                    df_sen = df_question[df_question.Sentiment == sen]
                    
                    if sen == 'POSITIVE':
                        clinic_pos = df_sen.Answer.sum()
                    elif sen == 'NEGATIVE':
                        clinic_neg = df_sen.Answer.sum()
                    else:
                        clinic_neu = df_sen.Answer.sum() 
                
            elif q == "Protocols":
                #protocol = df_question.Answer.sum()
                
                for sen in df_question.Sentiment.unique():
                
                    df_sen = df_question[df_question.Sentiment == sen]
                    
                    if sen == 'POSITIVE':
                        protocol_pos = df_sen.Answer.sum()
                    elif sen == 'NEGATIVE':
                        protocol_neg = df_sen.Answer.sum()
                    else:
                        protocol_neu = df_sen.Answer.sum() 
                
            elif q == "Cost":
                #cost = df_question.Answer.sum()
                #print(cost)
                for sen in df_question.Sentiment.unique():
                    
                    df_sen = df_question[df_question.Sentiment == sen]
                    
                    if sen == 'POSITIVE':
                        cost_pos = df_sen.Answer.sum()
                    elif sen == 'NEGATIVE':
                        cost_neg = df_sen.Answer.sum()
                    else:
                        cost_neu = df_sen.Answer.sum() 
                
            else:
                #wrong = df_question.Answer.sum()
                
                for sen in df_question.Sentiment.unique():
                
                    df_sen = df_question[df_question.Sentiment == sen]
                    
                    if sen == 'POSITIVE':
                        problem_pos = df_sen.Answer.sum()
                    elif sen == 'NEGATIVE':
                        problem_neg = df_sen.Answer.sum()
                    else:
                        problem_neu = df_sen.Answer.sum() 
                
        # Create one string for each topic
        
        list_of_reviews = [doctor_pos,doctor_neg,doctor_neu,advice_pos,advice_neg,advice_neu,nurse_pos,nurse_neg,nurse_neu,clinic_pos,clinic_neg,clinic_neu,protocol_pos,protocol_neg,protocol_neu,cost_pos,cost_neg,cost_neu,problem_pos,problem_neg,problem_neu]

                
        doc_score = df_question.avg_doc_score.mean()
        avg_doc_score=[]
        for i in range(21):
            avg_doc_score.append(doc_score)
        
        clinic_score = df_question.avg_clinic_score.mean()
        avg_clinic_score=[]
        for i in range(21):
            avg_clinic_score.append(doc_score)
        
        
        data = list(zip(clinic_name,topic_list,list_of_reviews,avg_clinic_score,avg_doc_score))
        
        df_new = pd.DataFrame(data, columns=['Clinic_name','Topics','Reviews','Avg_clinic_score','Avg_doc_score'])
    
        df_final = pd.concat([df_final,df_new], ignore_index=True)
    
    return df_final
            
```

```python
def concat_yelp_reviews(df):
    df_final = pd.DataFrame()
    for clinic in df.Clinic.unique():
        #print(clinic)
        clinic_name = [clinic,clinic,clinic,clinic,clinic,clinic,clinic]
        
        df_temp = df[df.Clinic == clinic]
        doctor=[]
        advice=[]
        nurse = []
        clinic = []
        protocol = []
        cost = []
        wrong = []
        
        topics = ["experience with doctor","advice give prospective patient","experience with nurse",
                 "experience with clinic","protocols and success","cost",'specific things went wrong']
        
        for q in topics:
            
            df_question = df_temp[df_temp.topics_predicted == q]
            
            
            if q == "experience with doctor":
                doctor = df_question.Reviews_by_sentence.sum()
                
            elif q == "advice give prospective patient":
                advice = df_question.Reviews_by_sentence.sum()
                
            elif q == "experience with nurse":
                nurse = df_question.Reviews_by_sentence.sum()
                
            elif q == "experience with clinic":
                clinic = df_question.Reviews_by_sentence.sum()
                
            elif q == "protocols and success":
                protocol = df_question.Reviews_by_sentence.sum()
                
            elif q == "cost":
                cost = df_question.Reviews_by_sentence.sum()
                #print(cost)
                
            else:
                if len(df_question)<1:
                    wrong = "Empty"
                
                else:
                    wrong = df_question.Reviews_by_sentence.sum()
                
        # Create one string for each topic
        
        list_of_reviews = [doctor,advice,nurse,clinic,protocol,cost,wrong]
        list_of_topics = ['doctor','advice','nurse','clinic','protocol','cost','What went Wrong']
        
        data = list(zip(clinic_name,list_of_topics,list_of_reviews))
        
        df_new = pd.DataFrame(data, columns=['Clinic_name','Topics','Reviews'])
    
        df_final = pd.concat([df_final,df_new], ignore_index=True)
    
    return df_final
```

# Summary of yelp reviews

```python
def yelp_summary(df):
    df['summary'] = np.nan
    summary_list = []
    n=0
    for r in df.Reviews:
        if '"' in r:
            r = r.replace('"',"")
            #print("replace")
        if "\xa0" in r:
            r = r.replace("\xa0", ". ")
        
        if len(r)>10000:
            sum_ = model(r,ratio=0.5)
            summary_list.append(sum_)
            print(n)
            n+=1
            
        elif len(r)>1000:
            sum_ = model(r,ratio=1)
            summary_list.append(sum_)
            print(n)
            n+=1
            
        elif len(r) > 0:
            sum_ = r
            summary_list.append(sum_)
            #print(n, "less than 1000")
            n+=1
            
        else:
            #print(n,"Empty")
            n+=1
            continue
    df.summary = summary_list
    return df
```
